```python
from cmdstanpy import CmdStanModel


import arviz as az
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
import pandas as pd

F = 'Filip'
L = 'Gąciarz'
N = (len(F) + len(L))*100
print(N)
```

```python
model_code_1=CmdStanModel(stan_file='/home/lab6_code_1.stan')
samples = model_code_1.sample(data={"N":N})
```

```python
model_code_2=CmdStanModel(stan_file='/home/lab6_code_2.stan')
samples_2 = model_code_2.sample(data={"N":N})
```

```python
df1 = samples.draws_pd()
df1
```

```python
df2 = samples_2.draws_pd()
df2
```

```python
y = [df1[f'y[{i}]'][0] for i in range(1,N+1)]
```

```python
model_code_3=CmdStanModel(stan_file='/home/lab6_code_3.stan')
samples_3 = model_code_3.sample(data={"N":N, 'y':y})
```

```python
df3 = samples_3.draws_pd()
df3
```

```python
model_code_4=CmdStanModel(stan_file='/home/lab6_code_4.stan')
samples_4 = model_code_4.sample(data={"N":N, 'y':y})
```

```python
df4 = samples_4.draws_pd()
df4
```

```python
compare_model_loo = az.compare(
    {"Normal model": az.from_cmdstanpy(samples_3),
    "Student model": az.from_cmdstanpy(samples_4)},
    ic="loo",
    scale="deviance")
compare_model_loo.head()
```

```python
compare_model_waic = az.compare(
    {"Normal model": az.from_cmdstanpy(samples_3),
    "Student model": az.from_cmdstanpy(samples_4)},
    ic="waic",
    scale="deviance")
compare_model_waic.head()
```

```python
fig, ax = plt.subplots(2,1)
az.plot_compare(compare_model_loo, insample_dev=True, ax = ax[0])
ax[0].set_title("loo criteria")
az.plot_compare(compare_model_waic, insample_dev=True, ax = ax[1])
ax[1].set_title("waic criteria")
plt.tight_layout()
plt.show()
```
