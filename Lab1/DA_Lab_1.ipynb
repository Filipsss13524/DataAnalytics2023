```python
ys = []
xs = []
alphas = []
betas = []
for i in range(1000):
    alpha = np.random.normal(2, 0.1)
    beta = np.random.normal(0.2, 0.2)
    x = np.random.normal(0, 3)
    sigma = halfnorm.rvs(loc = 0, scale = 1, size = 1)
    mu = beta*x + alpha
    y = np.random.normal(mu, sigma, 1000)
    xs.append(x)
    ys.append(y)
    alphas.append(alpha)
    betas.append(beta)
```

```python
plt.scatter(xs,ys[0])
for a,b in zip (alphas, betas):
  x = np.linspace(-6,6,100)
  y = b*x + a
  plt.plot(x,y,'r',alpha=0.05)
plt.xlim([-6,6])
plt.show()
```
